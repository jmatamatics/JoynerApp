```python
import fitz 
import fitz  # PyMuPDF

def extract_text_from_pdf(file_path):
    document = fitz.open(file_path)
    text = ''
    for page in document:
        text += page.get_text()
    return text

pdf_path = 'data/20240429091938234.pdf'
extracted_text = extract_text_from_pdf(pdf_path)
print(extracted_text)
```

```python
def image_to_text()
    import pytesseract
    from PIL import Image
    import fitz  # PyMuPDF
    text =''
    doc = fitz.open('data/20240429091938234.pdf')
    for page in doc:
        pix = page.get_pixmap()
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        text += pytesseract.image_to_string(img)
    #print("OCR Page {}: {}".format(page.number + 1, text))
        doc.close()
    return text
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def chunks(text):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=200)
    chunks = text_splitter.split_text(text)
    return chunks
```

```python

from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

def index(text_chunks):
    embeddings = OpenAIEmbeddings()
    vector = FAISS.from_texts(text_chunks, embeddings)
    #vector = FAISS.from_documents(text_chunks, embeddings)

    vector.save_local("faiss_index")
```
